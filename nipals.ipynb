```python
from numpy import linalg

def inv_diag(mtx):
    n = mtx.shape[0]
    return 1 / (mtx+np.ones(n) - np.eye(n) + np.eye(n)*1e-14) - np.ones(n) + np.eye(n)

def nipals(A,k):
    rows = A.shape[0]
    A = A - np.ones((rows,1)) * np.mean(A, axis=0)
    A = np.dot(A,inv_diag(np.diag(np.std(A, axis = 0))))
    
  
    initialvar = np.sum(np.var(A, axis=0))
    currentvar = initialvar
    T = np.ndarray(shape=(A.shape[0],k))
    P = np.ndarray(shape=(k,A.shape[1]))
    v = np.ndarray(shape=(1,k))
    for i in range(0, k):
        tk = A[:,k:k+1]
        q = False 
        while q == False:
            pk = np.dot(A.transpose(),tk)/np.dot(tk.transpose(),tk)
            pk = pk/np.linalg.norm(pk)
            tknew = np.dot(A,pk)/np.dot(pk.transpose(),pk)
            d = np.dot(tk.transpose(),tk)
            dnew = np.dot(tknew.transpose(),tknew)
            tk = tknew 
            if (dnew - d) <= 0.0001:
                q = True
                
        A = A - np.dot(tk,pk.transpose())
        T[:, i:i+1] =  tk
        P[i:i+1,:] =  pk.transpose() 
        oldvar = currentvar
        currentvar = np.sum(np.var(A, axis = 0))
        v[:, i:i+1] = (oldvar - currentvar)/initialvar
    return T, P, v

T, P, v = nipals(X,10)
```
